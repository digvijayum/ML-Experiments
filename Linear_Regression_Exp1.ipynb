```python
#Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Define and load the datast
#creates a DataFrame with Hours studied and corresponding Marks scored

data ={'Hours':[1,2,4,5,6],
       'Marks':[50,60,80,90,100]}
df = pd.DataFrame(data)
print(df)
```

```python
#Seperate input and output variables
X = df[['Hours']].values       #Independent variable as 2D array
Y = df[['Marks']].values      #Dependent variable as 2D array
```

```python
#Train and Linear Regression model

model = LinearRegression()     #create model bject
model.fit(X,Y)                 #Train the model using the data
```

```python
#model learns coefficients and output the line equation
slope = model.coef_[0][0]
intercept = model.intercept_[0]
print(f"Equation of line: y = {slope:2f}X + {intercept:.2f}")
```

```python
#Predict values and visualize

df['Predicted Marks'] = model.predict(X)
plt.scatter(df['Hours'], df['Marks'], color='blue', label='Actual Marks')
plt.plot(df['Hours'], df['Predicted Marks'], color='red', label='Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.title('Linear Regression: Hours vs Marks')
plt.legend()
plt.show()
```

```python
#Make predicition for new data

hours_to_predict = np.array([[7]])
predicted_marks = model.predict(hours_to_predict)[0][0]
print(f"Predicted marks for 3 hours of study:{predicted_marks}")
```
